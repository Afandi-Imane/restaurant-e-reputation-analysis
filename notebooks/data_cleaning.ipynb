```python
from datetime import datetime, timedelta
import re

def convertir_en_date(valeur):
    if pd.isna(valeur):
        return None
    
    valeur = valeur.lower().replace("modifié ", "")  # enlever "Modifié "
    maintenant = datetime.today()
    
    nombre = 1  # Par défaut "un" ou "une"
    unite = "jour"
    
    # Extraire nombre et unité
    match = re.search(r"il y a (\d+|un|une) (\w+)", valeur)
    if not match:
        return None
    
    nombre_str, unite = match.groups()
    
    # Convertir "un"/"une" en 1
    if nombre_str in ["un", "une"]:
        nombre = 1
    else:
        nombre = int(nombre_str)
    
    # Traduction en timedelta
    if "minute" in unite:
        date_finale = maintenant - timedelta(minutes=nombre)
    elif "heure" in unite:
        date_finale = maintenant - timedelta(hours=nombre)
    elif "jour" in unite:
        date_finale = maintenant - timedelta(days=nombre)
    elif "semaine" in unite:
        date_finale = maintenant - timedelta(weeks=nombre)
    elif "mois" in unite:
        date_finale = maintenant - pd.DateOffset(months=nombre)
    elif "an" in unite or "ans" in unite:
        date_finale = maintenant - pd.DateOffset(years=nombre)
    else:
        return None

    return date_finale.strftime("%d/%m/%Y")
```

```python
df = df.copy()
df['Date'] = df['Date'].apply(convertir_en_date)
```

```python

def extraire_intervalle(prix):
    if pd.isna(prix) or prix.strip() == "":
        return "0"

    prix = prix.strip()

    # Mapping pour les $
    dollar_intervalles = {
        "$": "50–70 MAD",
        "$$": "70–150 MAD",
        "$$$": "150–300 MAD",
        "$$$$": "300–500 MAD"
    }
    if prix in dollar_intervalles:
        return dollar_intervalles[prix]

    # Cas + de 500 MAD (ou variations)
    match_plus = re.search(r"\+\s*(de)?\s*(\d+)\s*MAD", prix, re.IGNORECASE)
    if match_plus:
        return f"+{match_plus.group(2)} MAD"

    # Cas 500 MAD
    match_unique = re.match(r"^(\d+)\s*MAD$", prix)
    if match_unique:
        return f"{match_unique.group(1)} MAD"

    # Cas intervalle : 100–500 MAD ou 100-500 MAD
    match_range = re.search(r"(\d+)[–\-](\d+)\s*MAD", prix)
    if match_range:
        return f"{match_range.group(1)}–{match_range.group(2)} MAD"

    return "0"

# Fonction pour calculer la moyenne de l’intervalle
def calculer_moyenne(intervalle):
    # +500 MAD → 500
    match_plus = re.match(r"\+(\d+)\s*MAD", intervalle)
    if match_plus:
        return int(match_plus.group(1))

    # 500 MAD → 500
    match_unique = re.match(r"^(\d+)\s*MAD$", intervalle)
    if match_unique:
        return int(match_unique.group(1))

    # 100–400 MAD
    match_range = re.match(r"(\d+)[–\-](\d+)\s*MAD", intervalle)
    if match_range:
        min_val = int(match_range.group(1))
        max_val = int(match_range.group(2))
        return (min_val + max_val) // 2

    return 0
# Appliquer les fonctions
df["Intervalle (MAD)"] = df["Prix"].apply(extraire_intervalle)
df["Prix moyen (MAD)"] = df["Intervalle (MAD)"].apply(calculer_moyenne)
```

```python
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from unidecode import unidecode

import string
import re
import pandas as pd
import emoji

# Liste des stop words français étendue
stop_words = set(stopwords.words('french'))
# Ajout de mots spécifiques à supprimer
stop_words.update(['kfc', 'casa', 'restaurant', '...', 'alors', 'etc', 'plus', 'tres','dhs'])

def enlever_emojis(texte):
    return emoji.replace_emoji(texte, replace='')  # supprime tous les emojis

def nettoyer_avis(texte):
    if pd.isnull(texte):
        return ""

    texte = enlever_emojis(texte)                  # Supprimer les emojis
    texte = texte.lower()                          # minuscules
    #texte = unidecode(texte)                       # accents
    texte = re.sub(r'[^\w\s]', ' ', texte)         # ponctuation
    texte = re.sub(r'\d+', ' ', texte)             # chiffres
    texte = re.sub(r'\s+', ' ', texte).strip()     # espaces multiples

    mots = texte.split()
    mots_utiles = [mot for mot in mots if mot not in stop_words]

    return ' '.join(mots_utiles)
```

```python
df['Commentaire_nettoye'] = df['Commentaire'].apply(nettoyer_avis)
```
